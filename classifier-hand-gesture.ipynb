```python
import os
import time
import tensorflow as tf 
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras.models import Sequential 
import csv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
X=pd.read_csv("/kaggle/input/gestures/gestures.csv")
X
```

```python
y=pd.DataFrame(X['Label'])
X.drop("Label",axis=1,inplace=True)
X
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X=pd.DataFrame(X_scaled,columns=X.columns)
X
```

```python

labels=['left_back','right_back','left_forward','right_forward',
        'left_thumbs_down','right_thumbs_down','left_thumbs_left','right_thumbs_left',
        'left_thumbs_right','right_thumbs_right','left_thumbs_up','right_thumbs_up']
one_hot_encoder = OneHotEncoder(categories=[labels],sparse=False)


y_one_hot = pd.DataFrame(one_hot_encoder.fit_transform(y.values.reshape(-1, 1)))
y_one_hot.columns=labels

y=y_one_hot
y_one_hot.head()
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

print("Training data shape:", X_train.shape, y_train.shape)
print("Testing data shape:", X_test.shape, y_test.shape)
```

```python
model=Sequential([Input(shape=(63,)),
                  Dense(128,activation='relu'),
                  Dense(64,activation='relu'),
                  Dense(128,activation='relu'),
                  Dense(64,activation='relu'),
                  Dense(32,activation='relu'),
                  Dense(12,activation="softmax")
                 ]
)
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
model.fit(X_train, y_train, 
          epochs=10,
          batch_size=40,
          validation_split=0.2,      
          verbose=1)    
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```
